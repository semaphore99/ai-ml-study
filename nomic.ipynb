```python
import pandas as pd
df = pd.read_csv('imdb_dataset_embeddings.csv')
```

```python
# apply PCA to reduce the dimensionality of the embeddings and then plot the embeddings
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
thing = pca.fit_transform(df['embeddings'].to_numpy().tolist())

```

```python
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# colour is coded by df['sentiment'], if sentiment is positive, colour is red, if sentiment is negative, colour is blue
ax.scatter(thing[:, 0], thing[:, 1], thing[:, 2], c=df['sentiment'].apply(lambda x: 'red' if x == 'positive' else 'blue'), marker='o')

plt.show()
```

```python
# use tensorflow to build a simple neural network to classify the sentiment of the reviews
import tensorflow as tf
from tensorflow.keras import layers
from tensorflow.keras import models
from sklearn.model_selection import train_test_split

# convert the sentiment column to a binary column
df['sentiment_binary'] = df['sentiment'].apply(lambda x: 1.0 if x == 'positive' else 0.0)

# split the data into training and testing data
x_train, x_test, y_train, y_test = train_test_split(df['embeddings'].to_numpy().tolist(), df['sentiment_binary'], test_size=0.2, random_state=42)

# create a simple neural network
model = models.Sequential()
model.add(layers.Dense(64, activation='relu', input_shape=(768,)))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

x_train = np.array(x_train)
y_train = np.array(y_train)

# train the model
model.fit(x_train, y_train, epochs=10, batch_size=32)
```

```python

# evaluate the model
x_test = np.array(x_test)
y_test = np.array(y_test)
model.evaluate(x_test, y_test)

# make predictions
y_pred = model.predict(x_test)

# convert the predictions to binary
y_pred_binary = [1 if x > 0.5 else 0 for x in y_pred]

# calculate the accuracy of the model
from sklearn.metrics import accuracy_score
print(f"accuracy score: {accuracy_score(y_test, y_pred_binary)}")

# calculate the confusion matrix
from sklearn.metrics import confusion_matrix
print(f"confusion matrix: {confusion_matrix(y_test, y_pred_binary)}")

# calculate the precision, recall and f1 score
from sklearn.metrics import precision_score, recall_score, f1_score
print(f"precision score: {precision_score(y_test, y_pred_binary)}")
print(f"recall score: {recall_score(y_test, y_pred_binary)}")
print(f"f1 score: {f1_score(y_test, y_pred_binary)}")
```

```python
# make 2 clusters from the embeddings using kmeans, use the training labels as cluster labels
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=2, random_state=0).fit(x_train)

# print the centroids of the clusters
print(kmeans.cluster_centers_)
# print the labels of the clusters
print(kmeans.labels_)
# print the inertia of the clusters
print(kmeans.inertia_)
# print the number of iterations of the clusters
print(kmeans.n_iter_)
# print the score of the clusters
print(kmeans.score(x_test))
```

```python
# for each row in x_test, predict the cluster
cluster_pred = kmeans.predict(x_test)

# calculate the accuracy of the cluster predictions
print(f"accuracy score: {accuracy_score(y_test, cluster_pred)}")
```

```python
# put the test data and predictions (both neural network and cluster predictions) into a dataframe
# each row in x_test should be a single column in the dataframe
data = [[list(row)] for row in x_test]
df_test = pd.DataFrame(data, columns=['embeddings'])
df_test['sentiment'] = y_test
df_test['sentiment_pred_neural'] = y_pred_binary
df_test['sentiment_pred_cluster'] = cluster_pred
```

```python
# compare accuracy of neural network and cluster predictions
print(f"accuracy score neural network: {accuracy_score(y_test, y_pred_binary)}")
print(f"accuracy score cluster: {accuracy_score(y_test, cluster_pred)}")

```
